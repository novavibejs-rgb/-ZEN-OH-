```python
# segunda parte do bot

def esperar(min=0.6, max=2.2):
    time.sleep(random.uniform(min, max))

def mover_mouse(driver, elemento):
    try:
        ActionChains(driver).move_to_element(elemento).perform()
        esperar(0.2, 0.8)
    except:
        pass

for i in times.keys():
    print(f"\r🔍 Procurando o time:     ➡➡➡{i}")
    try:
        # tenta encontrar a imagem do time
        imagem = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located(
                (By.XPATH, f'//img[contains(@alt,"{i}")]')
            )
        )
                                        
        # rola até o elemento ficar visível
        driver.execute_script(
            "arguments[0].scrollIntoView({block: 'center'});",
            imagem
        )

        print("✅ Time encontrado!")

        container = WebDriverWait(imagem, 10).until(
            lambda x: x.find_element(By.XPATH, './ancestor::div[contains(@id,"g_")]')
        )
        # pega o botão da estrela
        botao = container.find_element(
            By.CSS_SELECTOR,
            'button[data-testid*="favorite"]'
        )

        # garante que o botão está clicável
        WebDriverWait(container, 10).until(
            EC.element_to_be_clickable((By.CSS_SELECTOR, 'button[data-testid*="favorite"]'))
        )

        # clica no botão se não estiver favoritado
        if botao.get_attribute("aria-pressed") == "false":
            botao.click()
            print("⭐ Favoritado!")
        else:
            print("⭐ Já estava favoritado!")
            print()

    except TimeoutException:
        # se não encontrar, passa pro próximo time
        print(f"\r⚠ o time {i}, \n Nao Joga Hoje, 💥 proximo")
        print()
        continue

print("\n⚽:",len(times),"times na sua lista ✅")
# clicar no meus favorito, aguardando 10 seg
favorit = driver.find_element(By.XPATH,'/html/body/nav/div/a').click()
esperar(0.1,2.2)
driver.execute_script("window.scrollBy(0, 300);")

#muda abas
driver.switch_to.window(driver.window_handles[1])
time.sleep(3)

loc = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.ID, "container-main-left"))
)

botao = WebDriverWait(loc, 10).until(
    EC.element_to_be_clickable((By.XPATH, './/a[contains(@href,"soccer")]'))
)

mover_mouse(driver, botao)
esperar()
driver.execute_script("arguments[0].scrollIntoView({block: 'center'});", botao)
esperar(0.5, 1.5)

botao.click()
esperar(1.5, 3.5)

body = driver.find_element(By.TAG_NAME, "body")

# 🔄 scroll humano (não fixo)
for i in range(random.randint(8, 12)):
    body.send_keys(Keys.PAGE_DOWN)
    esperar(0.2, 0.7)

body.send_keys(Keys.HOME)
esperar()

wait = WebDriverWait(driver, 30)

for nome in times.values():
    print(f"\n🔍 Procurando: {nome}")

    try:
        encontrou = False
        xpath = f'.//div[normalize-space()="{nome}"]'

        wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, ".fixture-body")))

        jogos = driver.find_elements(By.CSS_SELECTOR, ".fixture-body")

        for i in range(len(jogos)):
            try:
                jogos = driver.find_elements(By.CSS_SELECTOR, ".fixture-body")
                jogo = jogos[i]

                mover_mouse(driver, jogo)
                driver.execute_script(
                    "arguments[0].scrollIntoView({block: 'center'});",
                    jogo
                )
                esperar(0.5, 1.5)

                elemento_time = jogo.find_element(By.XPATH, xpath)

                link = jogo.find_element(By.XPATH, ".//a")

                mover_mouse(driver, link)
                esperar(0.3, 1.0)

                driver.execute_script("arguments[0].click();", link)

                print("✅ Cliquei no jogo")
                encontrou = True
                break

            except NoSuchElementException:
                continue

            except Exception as e:
                print("🔄 Stale detectado, tentando novamente...")
                continue

        if not encontrou:
            print("❌ Não achou esse jogo")
            continue

        wait.until(EC.presence_of_element_located((By.TAG_NAME, "body")))
        esperar(1.5, 3.0)

        erro = driver.find_elements(By.CSS_SELECTOR, '[data-test="message-box-error-msg"]')

        if erro:
            print("⚠️ Sem dados, voltando...")
            driver.back()
            esperar(1.0, 2.5)

            wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, ".fixture-body")))
            continue

        try:
            botao_sim = wait.until(
                EC.element_to_be_clickable(
                    (By.XPATH,'//a[.//span[normalize-space()="Sim"]]'))
                    )

            mover_mouse(driver, botao_sim)
            esperar(0.3, 1.0)

            driver.execute_script("arguments[0].click();", botao_sim)
            print("✔️ Cliquei em SIM")

        except:
            print("⚠️ Botão SIM não encontrado")

        valor = wait.until(
            EC.element_to_be_clickable((By.XPATH, '//input[@placeholder="Montante"]'))
        )

        mover_mouse(driver, valor)
        esperar()

        valor.clear()

        # digitação humana
        for n in "1":
            valor.send_keys(n)
            time.sleep(random.uniform(0.05, 0.2))

        print("💰 Valor inserido")

        botao_apostar = wait.until(
        EC.element_to_be_clickable((By.ID, "betslip-placebet-btn")))
        mover_mouse(driver, botao_apostar)
        botao_apostar.click()
        print("🚀 realizando aposta")
        esperar(1.0,2.0)

        botao_limpar = wait.until(
            EC.element_to_be_clickable((By.CSS_SELECTOR, 'li.clear-bet-slip a'))
        )
        mover_mouse(driver, botao_limpar)
        esperar()

        botao_limpar.click()
        esperar(2.0, 4.0)

        print("🧹 Bilhete limpo")

        driver.back()
        esperar(1.0, 2.5)

        wait.until(
            EC.presence_of_element_located((By.CSS_SELECTOR, ".fixture-body"))
        )

    except Exception as e:
        print("🔥 Erro geral:", e)
        driver.back()
        esperar(1.0, 2.0)
        continue

botao_logo = wait.until(
    EC.element_to_be_clickable((By.CSS_SELECTOR, 'a[data-test="header-logo"]'))
)

mover_mouse(driver, botao_logo)
esperar()

botao_logo.click()

print("\n🏁 BOT FINALIZADO\n")
```
